```python
import pandas as pd

import os
```

```python
# downloadsディレクトリのパスを設定
downloads_dir = os.path.join(os.getcwd(), "downloads")

def get_year_list(downloads_dir):
    """downloads直下のすべてのサブディレクトリ（年と仮定）のリストを取得"""
    year_list = [d for d in os.listdir(downloads_dir) if os.path.isdir(os.path.join(downloads_dir, d))]
    return year_list

def get_csv_file_list(downloads_dir, year):
    """指定した年のサブディレクトリ内のすべてのCSVファイルのパスをリストに格納"""
    utf8_csv_dir = os.path.join(downloads_dir, year, "csv")
    csv_file_list = [
        os.path.join(utf8_csv_dir, file)
        for file in os.listdir(utf8_csv_dir)
        if file.endswith(".csv")
    ]
    return csv_file_list

# 年のリストを取得
year_list = get_year_list(downloads_dir)
print(year_list)

# 年を手動で選択（例として最初の年を選択）
selected_year = year_list[2]  # 適宜変更可能

# 選択した年のCSVファイルのパスを取得
csv_file_list = get_csv_file_list(downloads_dir, selected_year)

# 確認のためリストを出力
for csv_file in csv_file_list:
    print(csv_file)
```
